```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df = pd.read_csv("/kaggle/input/sleep-health-and-lifestyle-dataset/Sleep_health_and_lifestyle_dataset.csv")
df
```

```python
df.info()
```

```python
df = df.fillna('None')
df
```

```python
df.info()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(df['Age'], df['Sleep Disorder'], alpha=0.5, color='green')
plt.title('The relationship of age to sleep disorder')
plt.xlabel('Age')
plt.ylabel('Sleep Disorder')
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(df['Sleep Disorder'], df['Occupation'], alpha=0.5, color='green')
plt.title('The relationship of occupation to sleep disorder')
plt.xlabel('Occupation')
plt.ylabel('Sleep Disorder')
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(df['Gender'], df['Sleep Disorder'], alpha=0.5, color='green')
plt.title('The relationship of gender to sleep disorder')
plt.xlabel('Gender')
plt.ylabel('Sleep Disorder')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Sleep Duration', y='Quality of Sleep')
plt.title('Sleep Duration vs. Quality of Sleep')
plt.xlabel('Sleep Duration (hours)')
plt.ylabel('Quality of Sleep')
plt.show()
```

```python
average_sleep_by_occupation = df.groupby('Occupation')['Sleep Duration'].mean().reset_index()
print(average_sleep_by_occupation)

# Bar plot for Average Sleep Duration by Occupation
plt.figure(figsize=(12, 8))
sns.barplot(data=average_sleep_by_occupation, x='Occupation', y='Sleep Duration')
plt.title('Average Sleep Duration by Occupation')
plt.xlabel('Occupation')
plt.ylabel('Average Sleep Duration (hours)')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Physical Activity Level', y='Sleep Duration')
plt.title('Physical Activity Level vs. Sleep Duration')
plt.xlabel('Physical Activity Level')
plt.ylabel('Sleep Duration (hours)')
plt.show()

correlation_activity_sleep_duration = df['Physical Activity Level'].corr(df['Sleep Duration'])
print(f"Correlation between Physical Activity Level and Sleep Duration: {correlation_activity_sleep_duration}")
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Physical Activity Level', y='Quality of Sleep')
plt.title('Physical Activity Level vs. Quality of Sleep')
plt.xlabel('Physical Activity Level')
plt.ylabel('Quality of Sleep')
plt.show()

correlation_activity_sleep_quality = df['Physical Activity Level'].corr(df['Quality of Sleep'])
print(f"Correlation between Physical Activity Level and Quality of Sleep: {correlation_activity_sleep_quality}")
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Stress Level', y='Sleep Duration')
plt.title('Stress Level vs. Sleep Duration')
plt.xlabel('Stress Level')
plt.ylabel('Sleep Duration (hours)')
plt.show()
correlation_stress_sleep_duration = df['Stress Level'].corr(df['Sleep Duration'])
print(f"Correlation between Stress Level and Sleep Duration: {correlation_stress_sleep_duration}")
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Stress Level', y='Quality of Sleep')
plt.title('Stress Level vs. Quality of Sleep')
plt.xlabel('Stress Level')
plt.ylabel('Quality of Sleep')
plt.show()

correlation_stress_sleep_quality = df['Stress Level'].corr(df['Quality of Sleep'])
print(f"Correlation between Stress Level and Quality of Sleep: {correlation_stress_sleep_quality}")
```

```python
stress_sleep_by_occupation = df.groupby('Occupation')[['Stress Level', 'Quality of Sleep']].mean().reset_index()
print(stress_sleep_by_occupation)

# Bar plot for Stress Level by Occupation
plt.figure(figsize=(12, 8))
sns.barplot(data=stress_sleep_by_occupation, x='Occupation', y='Stress Level')
plt.title('Average Stress Level by Occupation')
plt.xlabel('Occupation')
plt.ylabel('Average Stress Level')
plt.xticks(rotation=45)
plt.show()

# Bar plot for Quality of Sleep by Occupation
plt.figure(figsize=(12, 8))
sns.barplot(data=stress_sleep_by_occupation, x='Occupation', y='Quality of Sleep')
plt.title('Average Quality of Sleep by Occupation')
plt.xlabel('Occupation')
plt.ylabel('Average Quality of Sleep')
plt.xticks(rotation=45)
plt.show()
```

```python
# Box plot for Sleep Duration by BMI Category
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='BMI Category', y='Sleep Duration')
plt.title('Sleep Duration by BMI Category')
plt.xlabel('BMI Category')
plt.ylabel('Sleep Duration (hours)')
plt.show()

# Box plot for Quality of Sleep by BMI Category
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='BMI Category', y='Quality of Sleep')
plt.title('Quality of Sleep by BMI Category')
plt.xlabel('BMI Category')
plt.ylabel('Quality of Sleep')
plt.show()
```

```python
sleep_disorder_bmi = df.groupby('BMI Category')['Sleep Disorder'].value_counts(normalize=True).unstack().fillna(0)
print(sleep_disorder_bmi)

sleep_disorder_bmi.plot(kind='bar', stacked=True, figsize=(12, 8))
plt.title('Sleep Disorder Prevalence by BMI Category')
plt.xlabel('BMI Category')
plt.ylabel('Proportion')
plt.legend(title='Sleep Disorder')
plt.show()
```

```python
# Convert Blood Pressure to numeric values (taking systolic pressure)
df[['Systolic BP', 'Diastolic BP']] = df['Blood Pressure'].str.split('/', expand=True).astype(int)

# Scatter plot for Systolic BP vs. Sleep Duration
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Systolic BP', y='Sleep Duration')
plt.title('Systolic Blood Pressure vs. Sleep Duration')
plt.xlabel('Systolic Blood Pressure (mmHg)')
plt.ylabel('Sleep Duration (hours)')
plt.show()

# Calculate correlation
correlation_bp_sleep_duration = df['Systolic BP'].corr(df['Sleep Duration'])
print(f"Correlation between Systolic Blood Pressure and Sleep Duration: {correlation_bp_sleep_duration}")

# Scatter plot for Systolic BP vs. Quality of Sleep
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Systolic BP', y='Quality of Sleep')
plt.title('Systolic Blood Pressure vs. Quality of Sleep')
plt.xlabel('Systolic Blood Pressure (mmHg)')
plt.ylabel('Quality of Sleep')
plt.show()

# Calculate correlation
correlation_bp_sleep_quality = df['Systolic BP'].corr(df['Quality of Sleep'])
print(f"Correlation between Systolic Blood Pressure and Quality of Sleep: {correlation_bp_sleep_quality}")

# Scatter plot for Heart Rate vs. Sleep Duration
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Heart Rate', y='Sleep Duration')
plt.title('Heart Rate vs. Sleep Duration')
plt.xlabel('Heart Rate (bpm)')
plt.ylabel('Sleep Duration (hours)')
plt.show()

# Calculate correlation
correlation_hr_sleep_duration = df['Heart Rate'].corr(df['Sleep Duration'])
print(f"Correlation between Heart Rate and Sleep Duration: {correlation_hr_sleep_duration}")

# Scatter plot for Heart Rate vs. Quality of Sleep
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Heart Rate', y='Quality of Sleep')
plt.title('Heart Rate vs. Quality of Sleep')
plt.xlabel('Heart Rate (bpm)')
plt.ylabel('Quality of Sleep')
plt.show()

# Calculate correlation
correlation_hr_sleep_quality = df['Heart Rate'].corr(df['Quality of Sleep'])
print(f"Correlation between Heart Rate and Quality of Sleep: {correlation_hr_sleep_quality}")
```

```python
bp_hr_sleep_disorder = df.groupby('Sleep Disorder')[['Systolic BP', 'Heart Rate']].mean().reset_index()
print(bp_hr_sleep_disorder)

# Bar plot for Systolic Blood Pressure by Sleep Disorder status
plt.figure(figsize=(10, 6))
sns.barplot(data=bp_hr_sleep_disorder, x='Sleep Disorder', y='Systolic BP')
plt.title('Systolic Blood Pressure by Sleep Disorder Status')
plt.xlabel('Sleep Disorder')
plt.ylabel('Systolic Blood Pressure (mmHg)')
plt.show()

# Bar plot for Heart Rate by Sleep Disorder status
plt.figure(figsize=(10, 6))
sns.barplot(data=bp_hr_sleep_disorder, x='Sleep Disorder', y='Heart Rate')
plt.title('Heart Rate by Sleep Disorder Status')
plt.xlabel('Sleep Disorder')
plt.ylabel('Heart Rate (bpm)')
plt.show()
```

```python
# Scatter plot for Daily Steps vs. Sleep Duration
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Daily Steps', y='Sleep Duration')
plt.title('Daily Steps vs. Sleep Duration')
plt.xlabel('Daily Steps')
plt.ylabel('Sleep Duration (hours)')
plt.show()

correlation_steps_sleep_duration = df['Daily Steps'].corr(df['Sleep Duration'])
print(f"Correlation between Daily Steps and Sleep Duration: {correlation_steps_sleep_duration}")

# Scatter plot for Daily Steps vs. Quality of Sleep
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Daily Steps', y='Quality of Sleep')
plt.title('Daily Steps vs. Quality of Sleep')
plt.xlabel('Daily Steps')
plt.ylabel('Quality of Sleep')
plt.show()

correlation_steps_sleep_quality = df['Daily Steps'].corr(df['Quality of Sleep'])
print(f"Correlation between Daily Steps and Quality of Sleep: {correlation_steps_sleep_quality}")
```

```python
sleep_disorder_counts = df['Sleep Disorder'].value_counts()
print(sleep_disorder_counts)
plt.figure(figsize=(10, 6))
sns.barplot(x=sleep_disorder_counts.index, y=sleep_disorder_counts.values)
plt.title('Counts of Sleep Disorders')
plt.xlabel('Sleep Disorder')
plt.ylabel('Count')
plt.show()
```

```python
sleep_disorder_impact = df.groupby('Sleep Disorder')[['Sleep Duration', 'Quality of Sleep', 'Systolic BP', 'Heart Rate']].mean().reset_index()
print(sleep_disorder_impact)

# Bar plot for Sleep Duration by Sleep Disorder status
plt.figure(figsize=(10, 6))
sns.barplot(data=sleep_disorder_impact, x='Sleep Disorder', y='Sleep Duration')
plt.title('Sleep Duration by Sleep Disorder Status')
plt.xlabel('Sleep Disorder')
plt.ylabel('Sleep Duration (hours)')
plt.show()

# Bar plot for Quality of Sleep by Sleep Disorder status
plt.figure(figsize=(10, 6))
sns.barplot(data=sleep_disorder_impact, x='Sleep Disorder', y='Quality of Sleep')
plt.title('Quality of Sleep by Sleep Disorder Status')
plt.xlabel('Sleep Disorder')
plt.ylabel('Quality of Sleep')
plt.show()

# Bar plot for Systolic Blood Pressure by Sleep Disorder status
plt.figure(figsize=(10, 6))
sns.barplot(data=sleep_disorder_impact, x='Sleep Disorder', y='Systolic BP')
plt.title('Systolic Blood Pressure by Sleep Disorder Status')
plt.xlabel('Sleep Disorder')
plt.ylabel('Systolic Blood Pressure (mmHg)')
plt.show()

# Bar plot for Heart Rate by Sleep Disorder status
plt.figure(figsize=(10, 6))
sns.barplot(data=sleep_disorder_impact, x='Sleep Disorder', y='Heart Rate')
plt.title('Heart Rate by Sleep Disorder Status')
plt.xlabel('Sleep Disorder')
plt.ylabel('Heart Rate (bpm)')
plt.show()
```
