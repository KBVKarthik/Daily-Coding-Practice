```python
import warnings
warnings.filterwarnings('ignore')

death_rates_df = pd.read_csv('/kaggle/input/death-rates-united-states/Death_rates.csv')
death_rates_suicide_df = pd.read_csv('/kaggle/input/death-rates-united-states/Death_rates_for_suicide__by_sex__race__Hispanic_origin__and_age__United_States.csv')

death_rates_df_info = death_rates_df.info()
death_rates_df_head = death_rates_df.head()

death_rates_suicide_df_info = death_rates_suicide_df.info()
death_rates_suicide_df_head = death_rates_suicide_df.head()

death_rates_df_info, death_rates_df_head, death_rates_suicide_df_info, death_rates_suicide_df_head
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the aesthetic style of the plots
sns.set_style("whitegrid")

# Distribution of Death Rates Over the Years
plt.figure(figsize=(12, 6))
sns.lineplot(data=death_rates_df, x='YEAR', y='ESTIMATE')
plt.title('Distribution of Death Rates Over the Years')
plt.xlabel('Year')
plt.ylabel('Death Rate Estimate')
plt.show()
```

```python
# Comparison of Death Rates by Age Categories
plt.figure(figsize=(12, 6))
sns.boxplot(data=death_rates_df, x='AGE', y='ESTIMATE')
plt.title('Comparison of Death Rates by Age Categories')
plt.xlabel('Age Category')
plt.ylabel('Death Rate Estimate')
plt.xticks(rotation=45)
plt.show()
```

```python
# Trends in Death Rates for Different Demographic Groups (Sex, Race, Hispanic Origin)
# Since the second dataset includes detailed demographic information, we will use it for this plot
plt.figure(figsize=(12, 6))
sns.lineplot(data=death_rates_suicide_df, x='YEAR', y='ESTIMATE', hue='STUB_NAME')
plt.title('Trends in Death Rates for Different Demographic Groups')
plt.xlabel('Year')
plt.ylabel('Death Rate Estimate')
plt.legend(title='Demographic Group', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
# Filtering the dataset to include only race-specific data
race_specific_df = death_rates_suicide_df[death_rates_suicide_df['STUB_NAME'].str.contains('White|Black|Asian|American Indian|Hispanic|Other')]

# Visualizing race-specific trends over the years
plt.figure(figsize=(12, 6))
sns.lineplot(data=race_specific_df, x='YEAR', y='ESTIMATE', hue='STUB_NAME', marker='o')
plt.title('Race-Specific Trends in Suicide Rates Over the Years')
plt.xlabel('Year')
plt.ylabel('Suicide Rate Estimate')
plt.legend(title='Race', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()

# Additional statistics for more in-depth analysis
race_stats = race_specific_df.groupby(['YEAR', 'STUB_NAME'])['ESTIMATE'].describe().unstack()

race_stats.head(10)
```

```python
# Extracting data for specific age groups
age_groups = ['Under 15', '15-24 years', '25-34 years', '35-44 years', '45-54 years', '55-64 years', '65-74 years', '75-84 years', '85 years and over']
age_group_df = death_rates_suicide_df[death_rates_suicide_df['AGE'].isin(age_groups)]

# Visualizing age-specific trends over the years
plt.figure(figsize=(12, 6))
sns.lineplot(data=age_group_df, x='YEAR', y='ESTIMATE', hue='AGE', marker='o')
plt.title('Age-Specific Trends in Suicide Rates Over the Years')
plt.xlabel('Year')
plt.ylabel('Suicide Rate Estimate')
plt.legend(title='Age Group', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()

# Additional statistics for more in-depth analysis
age_stats = age_group_df.groupby(['YEAR', 'AGE'])['ESTIMATE'].describe().unstack()

age_stats.head(10)
```

```python
# Filtering the dataset to include only Hispanic-specific data
hispanic_df = death_rates_suicide_df[death_rates_suicide_df['STUB_NAME'].str.contains('Hispanic')]

# Visualizing Hispanic-specific trends over the years
plt.figure(figsize=(12, 6))
sns.lineplot(data=hispanic_df, x='YEAR', y='ESTIMATE', hue='AGE', marker='o')
plt.title('Hispanic-Specific Trends in Suicide Rates Over the Years')
plt.xlabel('Year')
plt.ylabel('Suicide Rate Estimate')
plt.legend(title='Age Group', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()

# Additional statistics for more in-depth analysis
hispanic_stats = hispanic_df.groupby(['YEAR', 'AGE'])['ESTIMATE'].describe().unstack()

hispanic_stats.head(10)
```

```python
# Filtering the dataset to include only Hispanic-specific data
hispanic_age_df = death_rates_suicide_df[
    death_rates_suicide_df['STUB_NAME'].str.contains('Hispanic') & 
    death_rates_suicide_df['AGE'].isin(age_groups)
]

# Visualizing Hispanic-specific trends by age over the years
plt.figure(figsize=(12, 6))
sns.lineplot(data=hispanic_age_df, x='YEAR', y='ESTIMATE', hue='AGE', marker='o')
plt.title('Hispanic-Specific Trends in Suicide Rates by Age Over the Years')
plt.xlabel('Year')
plt.ylabel('Suicide Rate Estimate')
plt.legend(title='Age Group', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()

# Additional statistics for more in-depth analysis
hispanic_age_stats = hispanic_age_df.groupby(['YEAR', 'AGE'])['ESTIMATE'].describe().unstack()

hispanic_age_stats.head(10)
```

```python
# Identifying peak suicide rates for Hispanics
peak_hispanic_rates = hispanic_age_df[hispanic_age_df.groupby(['AGE'])['ESTIMATE'].transform(max) == hispanic_age_df['ESTIMATE']]

# Highlighting the peak suicide rates in the visualization
plt.figure(figsize=(12, 6))
sns.lineplot(data=hispanic_age_df, x='YEAR', y='ESTIMATE', hue='AGE', marker='o')
plt.scatter(peak_hispanic_rates['YEAR'], peak_hispanic_rates['ESTIMATE'], color='red', zorder=5)
for i in range(peak_hispanic_rates.shape[0]):
    plt.text(peak_hispanic_rates.iloc[i]['YEAR'], peak_hispanic_rates.iloc[i]['ESTIMATE'], 
             f"{peak_hispanic_rates.iloc[i]['ESTIMATE']:.1f}", 
             color='black', 
             fontsize=9, 
             ha='right')

plt.title('Hispanic-Specific Trends in Suicide Rates by Age with Peaks Highlighted')
plt.xlabel('Year')
plt.ylabel('Suicide Rate Estimate')
plt.legend(title='Age Group', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()

peak_hispanic_rates.head(10)
```
