```python
data = pd.read_csv('/kaggle/input/rsna-2024-lumbar-spine-degenerative-classification/train.csv')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming `data` is your DataFrame
def class_balance_analysis(columns):
    for column in columns:
        print(f"\nClass Balance for {column}:")
        print(data[column].value_counts(normalize=True) * 100)
        plt.figure(figsize=(10, 5))
        sns.countplot(x=data[column])
        plt.title(f'Class Balance of {column}')
        plt.xticks(rotation=90)
        plt.show()

# Analyze class balance for all relevant columns
columns_to_analyze = data.columns[1:]  # All columns except 'study_id'
class_balance_analysis(columns_to_analyze)

# Correlation Analysis
# Convert categorical data to numerical for correlation analysis
data_encoded = data.copy()
for column in data.columns[1:]:
    data_encoded[column] = data[column].astype('category').cat.codes

correlation_matrix = data_encoded.corr()

plt.figure(figsize=(15, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```
