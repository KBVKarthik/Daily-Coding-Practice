```python
# libraries
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# see data
csv = pd.read_csv('/kaggle/input/nissan-all-models-price-prediction-dataset/nissan-dataset.csv')
csv.head()
```

```python
# summary statistics

# age statistics
print("Age Statistics:")
print(f"Minimum Age: {csv['age'].min()}")
print(f"Maximum Age: {csv['age'].max()}")
print(f"Average Age: {csv['age'].mean()}")

# price statistics
print("\nPrice Statistics:")
print(f"Minimum Price: {csv['price'].min()}")
print(f"Maximum Price: {csv['price'].max()}")
print(f"Average Price: {csv['price'].mean()}")

# kilometers statistics
print("\nKilometers Statistics:")
print(f"Minimum km: {csv['km'].min()}")
print(f"Maximum km: {csv['km'].max()}")
print(f"Average km: {csv['km'].mean()}")
```

```python
# Correlation matrix
corr_matrix = csv[['age', 'price', 'km']].corr()

# heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Regression of Price on Age
plt.figure(figsize=(10, 6))
sns.regplot(x='age', y='price', data=csv, scatter_kws={'s':100}, line_kws={'color':'red'})
plt.title('Regression of Price on Age')
plt.xlabel('Age')
plt.ylabel('Price')
plt.show()
```

```python
# Count of Vehicles by Condition
plt.figure(figsize=(10, 6))
sns.countplot(x='condition', data=csv, palette='viridis')
plt.title('Count of Vehicles by Condition')
plt.xlabel('Condition')
plt.ylabel('Count')
plt.show()
```

```python
# Average Price by Condition
plt.figure(figsize=(10, 6))
sns.barplot(x='condition', y='price', data=csv, palette='viridis')
plt.title('Average Price by Condition')
plt.xlabel('Condition')
plt.ylabel('Average Price')
plt.show()
```
